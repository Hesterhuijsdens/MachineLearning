Assignment 2 a.) Sketch the trajectories, indicate stability of fixed points. <br>
$x1' = -x2 + x2^3 = 0$ <br>
$ x2 (-1 + x2) = 0$ <br>
$x2 = 0$ or $ x2^2-1 = 0,$ and therefore $ x2^2 = 1$ <br>
$x2 = 0, x2 = sqrt(1)$ or $ x2 = -sqrt(1)$

$x2' = -x1 + x1^3 = 0$
$= x1 (-1 + x1^2) = 0$
$x1^2 - 1 = 0$
$x1^2 = 1$
$x1 = 0, x1 = sqrt(1), x1 = -sqrt(1)$

put x1 in second derivative
$x'' = -1 + 3 * 0^2 = -1$. X=0 is stable.
$x'' = -1 + 3 * 1^2 = 2$. X=1 is unstable.
$x'' = -1 + 3 * (-1)^2 = 2$. X=-1 is unstable.

```python
# sketch the function itself
def plot(x):
    return -x + x**3


plt.figure(0)
x = linspace(-2, 2, 20)
plt.plot(x, plot(x))
plt.axhline(y=0, linestyle='--', c='grey')
plt.show()
```

```python
def f(x,y):
    return -y + y**3

def g(x,y):
    return -x + x**3

x1 = []
y1 = []
x2 = []
y2 = []
def sys(iv1, iv2, dt, time):
    x1.append(iv1)
    y1.append(iv2)
    x2.append(iv1 + 4)
    y2.append(iv2 + 4)
    for i in range(time):
        x1.append(x1[i] + (f(x1[i],y1[i])) * dt)
        y1.append(y1[i] + (g(x1[i],y1[i])) * dt)
        x2.append(x2[i] + (f(x2[i],y2[i])) * dt)
        y2.append(y2[i] + (g(x2[i],y2[i])) * dt)
    return x1, y1, x2, y2

sys(-3, -1, 0.01, 5)
fig = plt.figure(figsize=(15,15))
fig.subplots_adjust(wspace = 0.5, hspace = 0.3)
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.plot(x1, 'r-', label='x1')
ax1.plot(y1, 'b-', label='x2')
ax1.set_title("Dynamics in time (init_x1 = -3, init_x2 = -1)")
ax1.set_xlabel("time")
ax1.grid()
ax1.legend(loc='best')

ax2.plot(x1, y1, color="blue")
ax2.set_xlabel("x1")
ax2.set_ylabel("x2")  
ax2.set_title("Phase space (init_x1 = -3, init_x2 = -1)")
ax2.grid()

ax3.plot(x2, 'r-', label='x1')
ax3.plot(y2, 'b-', label='x2')
ax3.set_title("Dynamics in time (init_x1 = 1, init_x2 = 3)")
ax3.set_xlabel("time")
ax3.grid()
ax3.legend(loc='best')

ax4.plot(x2, y2, color="blue")
ax4.set_xlabel("x1")
ax4.set_ylabel("x2")  
ax4.set_title("Phase space (init_x1 = 1, init_x2 = 3)")
ax4.grid()
plt.show()
```

b.) Find the two centers
$x1' = 2x1x2 = 0$
$x1 = 0 v x2 = 0$ (zero factor principle)

$x2' = 1/4 - x1^2 + x2^2 = 0$
plug x1 we just found into this function
$= 1/4 - 0^2 + x2^2 = 0$
$x2^2 = -1/4$
$x2 = sqrt(-1/4) v -sqrt(-1/4)$
use radial rule: sqrt(-1/4) = sqrt(-1) sqrt(1/4) = 0.5i$
$x2 = 0.5i v -0.5i$

```python
def f(x,y):
    return 2*x*y
def g(x,y):
    return 1/4 -x**2 + y**2

x = []
y = []
def sys(iv1, iv2, dt, time):
    x.append(iv1)
    y.append(iv2)
    for i in range(time):
        x.append(x[i] + (f(x[i],y[i])) * dt)
        y.append(y[i] + (g(x[i],y[i])) * dt)
    return x, y

sys(10, 2, 0.01, 1000)

fig = plt.figure(figsize=(15,5))
fig.subplots_adjust(wspace = 0.5, hspace = 0.3)
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.plot(x, 'r-', label='x1')
ax1.plot(y, 'b-', label='x2')
ax1.set_title("Dynamics in time")
ax1.set_xlabel("time")
ax1.grid()
ax1.legend(loc='best')

ax2.plot(x, y, color="blue")
ax2.set_xlabel("x1")
ax2.set_ylabel("x2")  
ax2.set_title("Phase space")
ax2.grid()
plt.show()
```

Assignment 2 c.) Discuss the number of FPs, their location and their stability as you vary a.

```python
def eq1(x1, x2, a):
    return x2 + x1 * x2 + a * x1 * x2**2

def eq2(x1, x2):
    return -x1 -x1**2 + x2**2

x1 = []
x2 = []
x3 = []
x4 = []
y = []
def sys(iv1, iv2, dt, time):
    x1.append(iv1)
    x2.append(iv1)
    x3.append(iv1)
    x4.append(iv1)
    y.append(iv2)
    for i in range(time):
        x1.append(x[i] + (eq1(x1[i],y[i], a=-4)) * dt)
        x2.append(x[i] + (eq1(x2[i],y[i], a=-2)) * dt)
        x3.append(x[i] + (eq1(x3[i],y[i], a=0)) * dt)
        x4.append(x[i] + (eq1(x4[i],y[i], a=2)) * dt)
        y.append(y[i] + (eq2(x1[i],y[i])) * dt)
    return x, y

sys(10, 2, 0.01, 10)

fig = plt.figure(figsize=(15,15))
fig.subplots_adjust(wspace = 0.5, hspace = 0.3)
ax1 = fig.add_subplot(4,2,1)
ax2 = fig.add_subplot(4,2,2)
ax3 = fig.add_subplot(4,2,3)
ax4 = fig.add_subplot(4,2,4)
ax5 = fig.add_subplot(4,2,5)
ax6 = fig.add_subplot(4,2,6)
ax7 = fig.add_subplot(4,2,7)
ax8 = fig.add_subplot(4,2,8)

ax1.plot(x1, 'r-', label='x1_prime')
ax1.plot(y, 'b-', label='x2_prime')
ax1.set_title("Dynamics in time (a=-4)")
ax1.set_xlabel("time")
ax1.grid()
ax1.legend(loc='best')

ax2.plot(x1, y, color="blue")
ax2.set_xlabel("x1")
ax2.set_ylabel("x2")
ax2.set_title("Phase space (a=-4)")
ax2.grid()

ax3.plot(x2, 'r-', label='x1_prime')
ax3.plot(y, 'b-', label='x2_prime')
ax3.set_title("Dynamics in time (a=-2)")
ax3.set_xlabel("time")
ax3.grid()
ax3.legend(loc='best')

ax4.plot(x2, y, color="blue")
ax4.set_xlabel("x1")
ax4.set_ylabel("x2")
ax4.set_title("Phase space (a=-2)")
ax4.grid()

ax5.plot(x3, 'r-', label='x1_prime')
ax5.plot(y, 'b-', label='x2_prime')
ax5.set_title("Dynamics in time (a=0)")
ax5.set_xlabel("time")
ax5.grid()
ax5.legend(loc='best')

ax6.plot(x3, y, color="blue")
ax6.set_xlabel("x1")
ax6.set_ylabel("x2")
ax6.set_title("Phase space (a=0)")
ax6.grid()

ax7.plot(x4, 'r-', label='x1_prime')
ax7.plot(y, 'b-', label='x2_prime')
ax7.set_title("Dynamics in time (a=2)")
ax7.set_xlabel("time")
ax7.grid()
ax7.legend(loc='best')

ax8.plot(x4, y, color="blue")
ax8.set_xlabel("x1")
ax8.set_ylabel("x2")
ax8.set_title("Phase space (a=2)")
ax8.grid()
plt.show()
plt.show()

```

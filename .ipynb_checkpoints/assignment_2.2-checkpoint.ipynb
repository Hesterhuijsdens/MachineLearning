```python
from numpy import *
import matplotlib.pyplot as plt

# avoid overflow warnings
seterr(all="ignore")
```

The Van der Pol oscillator is given by ùë•¬® + ùê∂(ùë•^2 ‚àí 1)ùë•Àô + ùúî2ùë• = 0 and has a limit
cycle. <br>
a) Show using a phase plane analysis (plot the null clines) that there is a limit
cycle

```python
def f(x,y):
    return y


def g(x,y):
    C = 1
    w = 1
    return -C * (x**2 - 1) * y - w**2 * x

x = []
y = []
def sys(iv1, iv2, dt, time):
    x.append(iv1)
    y.append(iv2)
    for i in range(time):
        x.append(x[i] + (f(x[i],y[i])) * dt)
        y.append(y[i] + (g(x[i],y[i])) * dt)
    return x, y

sys(0.00001, 0.00001, 0.01, 10000)

fig = plt.figure(figsize=(15,5))
fig.subplots_adjust(wspace = 0.5, hspace = 0.3)
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.plot(x, 'r-', label='x1')
ax1.plot(y, 'b-', label='x2')
ax1.set_title("Dynamics in time")
ax1.set_xlabel("time")
ax1.grid()
ax1.legend(loc='best')

ax2.plot(x, y, color="blue")
ax2.set_xlabel("x1")
ax2.set_ylabel("x2")  
ax2.set_title("Phase space")
ax2.grid()
plt.show()
```

```python
def modifiedEuler(f, x, h, params=None, t=0):
    xhat = x + h * f(x,t,params)
    return x + h * 0.5 * (f(x,t,params) + f(xhat,t,params))

def vanDerPolOscillator(x,t,params):
    dx_0 = x[1]
    dx_1 = -params[0]*(x[0]**2 - 1) * x[1] - x[0]*params[1]**2
    return asarray([dx_0,dx_1])
```

```python
# In the plot below, it can be noted that amplitude is influenced by omega.

# initialize variables
w = arange(0.0, 6.0, 2.0)
C = 1.0
t0 = 0.0
tf = 20.0
h = 0.05

X = zeros((len(w), int(tf/h)))
Y = zeros((len(w), int(tf/h)))

# iterate for every omega value
for i in range(len(w)):
    
    t = t0 + h
    params = asarray([C, w[i]])
    x0 = asarray([0.1, 0.0])
    x = x0
    X[i,0],Y[i,0] = x0
    step = 1

    # 20 time points
    while t < tf:
        x = modifiedEuler(vanDerPolOscillator, x, h, params=params)
        X[i, step] = x[0]
        Y[i, step] = x[1]
        t += h
        step += 1


plt.figure(0)
for i in range(len(w)):
    plt.plot(X[i],Y[i])

plt.xlabel("x")
plt.ylabel("dx/dt")
plt.legend(w)
plt.show()
```

```python
# In the plot below, it can be noted that amplitude is influenced by C.

# initialize variables
w = 1.0
C = arange(0.0, 6.0, 2.0)
t0 = 0.0
tf = 20.0
h = 0.05

X = zeros((len(C), int(tf/h)))
Y = zeros((len(C), int(tf/h)))

# iterate for every C value
for i in range(len(C)):
    
    t = t0 + h
    params = asarray([C[i], w])
    x0 = asarray([0.1, 0.0])
    x = x0
    X[i,0],Y[i,0] = x0
    step = 1

    # 20 time points
    while t < tf:
        x = modifiedEuler(vanDerPolOscillator, x, h, params=params)
        X[i, step] = x[0]
        Y[i, step] = x[1]
        t += h
        step += 1


plt.figure(0)
for i in range(len(C)):
    plt.plot(X[i],Y[i])

plt.xlabel("x")
plt.ylabel("dx/dt")
plt.legend(C)
plt.show()
```
